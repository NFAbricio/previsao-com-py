```python
# importando base de dados que a inteligencia ira analisar 

import pandas as pd

tabela = pd.read_csv("clientes.csv")
display(tabela)
display(tabela.info())
```

```python
# transformando os "object" em "int" para ser mais facil a leitura 
from sklearn.preprocessing import LabelEncoder

codificador = LabelEncoder()
tabela["profissao"] = codificador.fit_transform(tabela["profissao"])
tabela["mixcredito"] = codificador.fit_transform(tabela["mix_credito"])
tabela["comportamento_pagamento"] = codificador.fit_transform(tabela["comportamento_pagamento"])

display(tabela.info())
```

```python
#preparando a os testes e os treinos 

x = tabela.drop(["score_credito", "id_cliente"], axis=1)
y = tabela["score_credito"]

from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, test_size=0.2)
```

```python
# Importando os modelos de treinamento para nossa inteligencia artificial 

from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier 

modelo_arvore = RandomForestClassifier()
modelo_neighbor = KNeighborsClassifier()

modelo_arvore.fit(x_treino, y_treino)
modelo_neighbor.fit(x_treino, y_treino)
```
